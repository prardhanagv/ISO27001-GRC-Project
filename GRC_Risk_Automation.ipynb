```python
import csv

# Map Likelihood / Impact to numbers
level_map = {"Low": 1, "Medium": 2, "High": 3}

# Map numeric risk to Risk Level
def calc_risk_level(likelihood, impact):
    score = level_map[likelihood] * level_map[impact]
    if score <= 2:
        return "Low"
    elif score <= 4:
        return "Medium"
    else:
        return "High"

# Read CSV and update Risk Level
input_file = "risk_register.csv"
output_file = "risk_report.csv"


with open(input_file, mode='r', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = []
    for row in reader:
        risk_level = calc_risk_level(row['Likelihood'], row['Impact'])
        row['Risk_Level'] = risk_level
        rows.append(row)

# Write updated CSV
with open(output_file, mode='w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print(f"Risk report generated: {output_file}")


```

```python
# Step 1: Upload your Excel file
from google.colab import files
import pandas as pd

print("Please upload your risk_report.xlsx file")
uploaded = files.upload()  # This will open a file picker

# Step 2: Load the Excel file
input_file = list(uploaded.keys())[0]  # gets the uploaded file name
output_file = "risk_report_updated.xlsx"

df = pd.read_excel(input_file)

# Step 3: Map Likelihood / Impact to numbers
level_map = {"Low": 1, "Medium": 2, "High": 3}

def calc_risk_level(likelihood, impact):
    score = level_map[likelihood] * level_map[impact]
    if score <= 2:
        return "Low"
    elif score <= 4:
        return "Medium"
    else:
        return "High"

# Step 4: Calculate Risk Level for each row
df['Risk_Level'] = df.apply(lambda row: calc_risk_level(row['Likelihood'], row['Impact']), axis=1)

# Step 5: Print summary of High / Medium / Low risks
summary = df['Risk_Level'].value_counts()
print("\nRisk Summary:")
print(summary)

# Step 6: Save updated Excel file
df.to_excel(output_file, index=False)
print(f"\nUpdated risk report saved as: {output_file}")

# Step 7: Download updated file
files.download(output_file)
```

```python
from google.colab import files
import pandas as pd

# Upload your Excel file
print("Please upload your risk_register.xlsx file")
uploaded = files.upload()
input_file = list(uploaded.keys())[0]
output_file = "risk_report_clean.xlsx"

# Read Excel with correct headers (row 1)
df = pd.read_excel(input_file, header=1)

# Rename columns to standard names
df.columns = ['Risk_ID','Asset','Threat','Vulnerability','Likelihood','Impact','Risk_Level','Existing_Control','Recommended_Control']

# Drop completely empty rows
df = df.dropna(how='all')

# Map Likelihood / Impact to numbers
level_map = {"Low": 1, "Medium": 2, "High": 3}

def calc_risk_level(likelihood, impact):
    score = level_map.get(likelihood,0) * level_map.get(impact,0)
    if score <= 2:
        return "Low"
    elif score <= 4:
        return "Medium"
    else:
        return "High"

# Calculate Risk Level
df['Risk_Level'] = df.apply(lambda row: calc_risk_level(row['Likelihood'], row['Impact']), axis=1)

# Print summary of High/Medium/Low
summary = df['Risk_Level'].value_counts()
print("\nRisk Summary:")
print(summary)

# Save updated Excel
df.to_excel(output_file, index=False)
print(f"\nUpdated risk report saved as: {output_file}")

# Download the clean file
files.download(output_file)

```
